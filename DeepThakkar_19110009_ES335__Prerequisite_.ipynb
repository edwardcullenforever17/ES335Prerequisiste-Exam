1. How many multiplications and additions do you need to perform a matrix multiplication between a (n, k) and (k, m) matrix? Explain.

Solution:

Let A be a (n, k) matrix and B be a (k, m) matrix. Thus A has n rows and k columns and B has k rows and m columns. The resultant matrix R will be a (n, m) matrix.

To get any element of the R matrix, say $ r_{n_i,m_j} $ can be got by multiplying elements in the n_i row in A matric with the m_j coumn in B matrix. Each row in the A matrix has k elements and each column in B matric has k elements. Thus to get every $ r_{n_i,m_j} $, k multiplications are required and (k-1) additions are required. Thus to get all the terms of the matrix R a total of  **mnk mutliplications** and **mn(k-1) additions**.

Using this method, what is the time complexity of finding the median of the array? Can you suggest a better method? Can you implement both these methods in Python and compare against numpy.median routine in terms of time?

Solution:

**For the given method:**

Step 1: Finding the length of the array (n) --------------------------------------------------------------------------------------------Time Complexity: O(1)

Step 2: Create a counter (c)and give it the intial value n/2 -------------------------------------------------------------------Time Complexity: O(1)

Step 3: Iterating to finding the maximum element in the array------ -------------------------------------------------------Repeat n/2 times
> Step 3(i): Check if  c > 0 ------------------------------------------------------------------------------------------------------------------Time Complexity: O(1)

> Step 3(ii): Pass through the array to find the maximum element---------------------------------------------------Time Complexity: O(n)

> Step 3(iii): Delete the maximum element and decrease 1 from counter initiaed to l/2------------------Time Complexity: O(1)

Time complexity for finding the median of this array is O($n^2$)

**A Better Method**

Step 1: Finding the length of the array (n) --------------------------------------------------------------------------------------------Time Complexity: O(1)

Step 2: Sort the array using mergesort algorithm --------------------------------------------------------------------------------Time Complexity: O(nlogn)

Step 3: Find the median term = $({n/2})^{th}$ term -------------------------------------------------------------------------------------Time Complexity: O(n)

```python
## Taking the array as input
Aq3 = np.array([9, 8, 7, 6, 5, 4, 3, 2, 1])
print(Aq3)
```

```python
## Using the given Method
A1 = Aq3
l = len(A1)

for j in range(int(l/2)+1):
  l1 = len(A1)
  Max = 0
  for i in range(l1):
    if A1[i]> Max:
       Max = A1[i]
       i = i +1
    else:
      i = i +1
  k = np.where(A1 == Max)
  j = j - 1
  A1 = np.delete(A1,k)
print(Max)
```

```python
## A better method

A2 = Aq3
l = int(len(A2)/2)
A2s = np.sort(A2)
Med = A2s[l]
print(Med)
```

```python
## Using numpy.median routine

Med3 = np.median(Aq3)
print(Med3)
```

Thus the median obtained by all the three methods is consistent and is equal to 5 for the chosen array.

4. What is the gradient of the following function with respect to x and y?
\begin{align}
 x^2y + y^3sin(x)
\end{align}

Solution:

Gradient of a function is given by the formula,
\begin{align}
Grad(f(x,y)) = ∇ f(x,y) =  \frac{∂f}{∂x}i + \frac{∂f}{∂y}j
\end{align}
For
\begin{align}
f(x) =  x^2y + y^3sin(x)
\end{align}

Gradient with respect to x $ = \frac{∂ (x^2y + y^3sin(x))}{∂x} = 2xy + y^3cos(x) $

Gradient with respect to y $ = \frac{∂ (x^2y + y^3sin(x))}{∂y} = x^2 + 3y^2sin(x) $

10. Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as “a”, “b”, “c”, “d”, “e”. Find the column with the highest standard deviation. Find the row with the lowest mean.

Solution:

```python
import random
import pandas as pd

Mat = np.zeros((20, 5))

for i in range(20):
  for j in range(5):
    Mat[i][j] = Mat[i][j] + random.uniform(1,2)

DF1 = pd.DataFrame(Mat, columns = [ 'a', 'b', 'c', 'd', 'e'])

SD = np.std(DF1)
print(SD)

mini = np.minimum(SD[][1])
print(mini)
```
